## Cleaning our Dataset

```python
## creating a function to calculate the % of employees that did not input values for education & previous year rating 
## to enable us clean our data

def percent(x):
    r,c = data.shape ## data is the data frame
    perc = float((x/r)*100)
    return perc
```

```python
## The percentage of employees without record of education
## 4% of employees are without educational records

percent(data.education.isnull().sum())
```

```python
## The percentage of employees without record of previous year rating
## 7% of employees are without previous year rating

percent(data.previous_year_rating.isnull().sum())
```

```python
## We have to drop gender&recruitment_channel field to remove bias from our deductions

data.drop(columns=['gender','recruitment_channel'], inplace=True)
```

```python
## Since the percentage of people without previous year rating is relatively low, rather than dropping the field,
## we may substitute the null values with a measure of central tendency: mean,median or mode
## We will group the department&Region and substitute the measure of central tendency for the corresponding department&region in the null record

data.groupby(['department','region'])['previous_year_rating'].mean()  ## Getting the mean based on department and region
```

```python
## We need to determine if the mean is the best measure of central tendency to use
## It will if there are no outliers in the system and we check that with the aid of a boxplot

data.boxplot(column=['previous_year_rating'],grid=False);

## Since there is an outlier, we will substitute null values with their median
```

```python
data.groupby(['department','region'])['previous_year_rating'].median()  ## Getting the median based on department and region
```

```python
data['previous_year_rating'].fillna(data.groupby(['department','region'])['previous_year_rating'].transform('median'), inplace = True)  ##subs median value
```

## 4) Why did the Analytics Team perform better than the Procurement and Technology Team despite having lesser amount of Trainings

```python
data.head()
```

```python
## Lets Check if this was as a result of the educational background of Staff in the team
## First we are seperate 
data_new=data.groupby(['department','education']).count()[['employee_id']]
data_new.rename(columns={'employee_id':'No_of_Staff'}, inplace=True)
data_new
```

```python
## Confirming if the new data sets correlates with our original data set

data_new.sum() == data.shape[0]
```

```python
## We are going to define a function to calculate the percentage of staff in each department with respective educational qualification

def get_percent(x,y):   ### Where x is the department and y is the educational qualification
    A = (data_new.loc[x].loc[y]/data_new.loc[x].sum())*100
    return A
```

### For Analytics Department

```python
AB=get_percent("Analytics","Bachelor's")
AB
```

```python
AM=get_percent("Analytics","Master's & above")
AM
```

```python
ax=pd.Series({"Bachelors":AB,"Masters & above":AM}).astype('float').plot(kind='pie', radius=1.5,autopct='%1.1f%%',explode=[0.05,0.05])
ax.yaxis.set_visible(False)
plt.title('Analytics Department',pad=50,size=15);
```

### For Procurement Department
